# Data Wrangling
8.1 Column Handling

Irrelevant columns such as description were removed, column names were standardized, and columns were reordered for better structure. This improves readability, reduces unnecessary data complexity, and ensures a clean and organized dataset for analysis.

```python
# Rename columns
df.columns = df.columns.str.lower()

```

8.2 Feature Creation

A new feature content_age was created from the release year to measure how old each title is. This helps in understanding content distribution over time and enhances analytical insights.

```python
# Create content age (current year - release year)
df['content_age'] = 2026 - df['release_year']
```

8.3 Date Feature Extraction
The date_added column was converted to datetime format and components like year and month were extracted. This enables time-based analysis and helps identify trends in content additions.


```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

8.4 Data Type Conversion

Columns were converted to appropriate data types such as integer, category, and datetime. This ensures accurate analysis, improves memory efficiency, and enhances computational performance.

```python
df['type'] = df['type'].astype('category')
df['rating'] = df['rating'].astype('category')
```

# Data Cleaning
9.1 Duplicate Handling

```python
duplicates = df.duplicated().sum()
print("Duplicates:", duplicates)

df.drop_duplicates(inplace=True)
```

9.2 Missing Value Treatment

```python
df.isnull().sum()
```

# Data Visualization
Univariate Analysis

1. Content Type Distribution
(Count Plot)

The count plot displays the distribution of content types, showing the proportion of Movies and TV Shows available on the platform.

```python
sns.countplot(x='type', data=df)
plt.title("Content Type Distribution")
plt.show()
```

2. Release Year Distribution (Histogram)

The histogram shows how content is distributed across different release years, highlighting periods of higher production activity.


```python
plt.hist(df['release_year'], bins=30)
plt.title("Release Year Distribution")
plt.show()
```

3. Rating Distribution (Count Plot)

The chart illustrates the frequency of different age ratings assigned to the content.

```python
sns.countplot(y='rating', data=df)
plt.title("Rating Distribution")
plt.show()
```

4. Top 10 Countries (Bar Chart)

The bar chart presents the top 10 countries contributing the highest number of titles.

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()
```

5. Content Age Distribution (Histogram)

This histogram shows the distribution of content based on how old the titles are since their release year


```python
sns.histplot(df['content_age'], bins=20)
plt.title("Content Age Distribution")
plt.show()
```

# Bivariate Analysis
6. Type vs Rating (Count Plot with Hue)

The chart compares how ratings are distributed between Movies and TV Shows.

```python
plt.figure(figsize=(10,6))

sns.countplot(x='type', hue='rating', data=df)

plt.title("Content Type vs Rating")
plt.xlabel("Type")
plt.ylabel("Count")

plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

7. Release Year vs Type (Box Plot)

The box plot compares the release year distribution between Movies and TV Shows, highlighting variation and median trends.

```python
sns.boxplot(x='type', y='release_year', data=df)
plt.show()
```

8. Rating vs Content Type (Stacked Bar Chart)

This stacked bar chart shows the distribution of Movies and TV Shows across different rating categories, highlighting how audience ratings vary by content type.

```python
rating_type = pd.crosstab(df['rating'], df['type'])

rating_type.plot(kind='bar', stacked=True, figsize=(10,6))

plt.title("Rating vs Content Type")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

9. Movies by Country (Count Plot)

The chart displays the number of titles produced by different countries, emphasizing leading contributors.

```python
sns.countplot(y='country', data=df,
              order=df['country'].value_counts().iloc[:10].index)
plt.show()
```

10. Rating vs Release Year (Box Plot)

The box plot shows how release years vary across different content ratings.

```python
sns.boxplot(x='rating', y='release_year', data=df)
plt.xticks(rotation=90)
plt.show()
```

# Multivariate Analysis

11. Correlation Heatmap

The heatmap visualizes the correlation between numerical variables in the dataset.

```python
corr = df.select_dtypes(include=['int64','float64']).corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

12. Pairplot
The pair plot shows relationships and distributions among numerical variables in the dataset.

```python
sns.pairplot(df.select_dtypes(include=['int64','float64']))
plt.show()
```

13. Type vs Year Added (Count Plot with Hue)

The chart compares how Movies and TV Shows were added across different years.

```python
sns.countplot(x='year_added', hue='type', data=df)
plt.xticks(rotation=90)
plt.show()
```

14. Release Year Trend by Type (Line Plot)

The line chart shows the yearly trend of content releases separately for Movies and TV Shows.

```python
release_year_type_counts = df.groupby(['release_year', 'type']).size().reset_index(name='count')
sns.lineplot(x='release_year', y='count', hue='type', data=release_year_type_counts)
plt.title("Release Year Trend by Type")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

15. Violin Plot (Release Year by Type)

The violin plot visualizes the distribution and density of release years for Movies and TV Shows.

```python
sns.violinplot(x='type', y='release_year', data=df)
plt.show()
```

16. Country vs Type (Bar Chart)

The bar chart compares content type distribution across different countries.

```python
pd.crosstab(df['country'], df['type']).head(10).plot(kind='bar')
plt.show()
```

17. Month Added Distribution (Count Plot)

The chart shows the distribution of content additions across different months.

```python
sns.countplot(x='month_added', data=df)
plt.show()
```

18. Content Age vs Type (Box Plot)

The box plot compares how old Movies and TV Shows are in terms of release year.

```python
sns.boxplot(x='type', y='content_age', data=df)
plt.show()
```

19. Top 10 Genres (Bar Chart)

The bar chart displays the most common genres available on the platform.

```python
df['listed_in'].value_counts().head(10).plot(kind='barh')
plt.show()
```

20. Interactive Release Year Distribution (Plotly Histogram)

The interactive histogram shows the distribution of release years with separate color distinction for Movies and TV Shows.

```python
fig = px.histogram(df, x='release_year', color='type')
fig.show()
```

## Key Insights Summary


*   Movies constitute a larger portion of the content compared to TV Shows on the platform.
*  The majority of content was released after the year 2000, with significant growth observed after 2015.
*   The United States contributes the highest number of titles among all countries.
*   Most content falls under TV-14 and TV-MA ratings, indicating a focus on teenage and adult audiences.
*   A rapid increase in content addition was observed in recent years, showing platform expansion.
*   Certain genres dominate the platform, indicating audience preference trends.

## Conclusion

The Exploratory Data Analysis reveals that Amazon Prime has a strong dominance of movies over TV shows and focuses primarily on modern content released after 2000. The platform shows rapid growth in recent years, with significant contributions from specific countries and a concentration on particular audience ratings and genres.

Overall, the dataset reflects strategic content expansion and a focus on adult and contemporary entertainment.
